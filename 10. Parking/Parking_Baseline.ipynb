데이터 EDA  
=====
단지코드의 경우 동일 단지코드의 전용면적별 세대수의 합이 각 열의 총세대수와 일치 또한 공가수가 각 전용면적의 공가수가 아닌 단지의 공가 수로 보이므로 전용면적이 다른 경우가 있어서 각 열의 공가수가 적절한 데이터인지 의문이 듬.

기본데이터 탐색시 노선 수 및 버스 정류장 수의 데이터에 누락 존재 이 값을 어떻게 채울 것인가?

1. 임대건물구분이 상가인 겯우  
 1-1) 상가인 경우 임대보증금 및 임대료가 기입되지 않음 ->이를 어떻게 해결할 것인가?  
 1-2) 상가인 경우 공급유형이 임대상가로 전부 일치  
 1-3) 상가인 경우 자격유형이 D만 존재  

2. 임대건물구분이 아파트인 경우
 2-1) 지역 건물들이 상가보다는 다양하게 분포 (상가의 경우 : 강원도 충청남도 경상남도 대전광역시 부산광역시 제주특별자치도만 존재)  
 2-2) 공급 유형이 상가대비 많으며 일부 레이블의 경우 기간 차이만 존재 (예: 공공임대 5년 10년 50년 분납)
 2-3) 자격유형이 A부터 O까지 존재
 2-4) 암대보증금과 임대료에 '-'로 데이터가 들어있는 데이터가 존재 또한 NaN 값도 존재
 확인결과 총 3개의 단지에서 데이터가 '-'로 입력되어 있음. C1326, C1786, C2186
